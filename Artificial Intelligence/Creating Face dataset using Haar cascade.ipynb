```python
import cv2
import imutils
import time
import csv
import os
```

```python
cascade = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
detector = cv2.CascadeClassifier(cascade)
```

```python
Name = str(input("Enter your Name :"))
Roll_no = int(input("Enter you Roll No :"))
dataset = 'dataset'
sub_data = Name
path = os.path.join(dataset,sub_data)
if not os.path.isdir(path):
    os.mkdir(path)
    print(sub_data)
info = [str(Name),str(Roll_no)]
with open('student.csv','a') as csvFile :
    write = csv.writer(csvFile)
    write.writerow(info)
csvFile.close()

print("Video is Streaming :")
cam = cv2.VideoCapture(0)
time.sleep(1)
total = 0
while(total < 50) :
    print(total)
    frame = cam.read()[1]
    img = imutils.resize(frame,width = 400)
    rects = detector.detectMultiScale(cv2.cvtColor(img,cv2.COLOR_BGR2GRAY),scaleFactor = 1.1,minNeighbors=5,minSize = (30,30))
    
    for(x,y,width,height) in rects :
        cv2.rectangle(frame,(x,y),(x+width,y+height),(0,255,0),2)
        p = os.path.sep.join([path,"{}.png".format(str(total).zfill(5))])
        cv2.imwrite(p,img)
        total = total+1
    cv2.imshow("Frame",frame)
    key = cv2.waitKey(1)& 0xFF
    if(key == ord("e")):
        break
cam.release()
cv2.destroyAllWindows()
```
